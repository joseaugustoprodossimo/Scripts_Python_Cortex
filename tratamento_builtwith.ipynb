```python
##############################################################
#        Script: Tratamento dos Arquivos do Builtwith        #
#                    Developed By: Gus                       #
#        E-mail: gustavo.lima@cortex-intelligence.com        #
##############################################################


#Imports
import csv
import pandas as pd

#Pasta de Entrada;
input_file_path = r'C:\Projetos\Python\BuiltWith\websites-matrixcsv\\'

#Input Arquivos;
input_mx = 'mx' + '.csv'                   #Provedores de E-mail
input_analytics = 'analytics' + '.csv'     #Provedores de Analytics
input_cms = 'cms' + '.csv'                 #Provedores de CMS
input_framework = 'framework' + '.csv'     #Provedores de Framework
input_hosting = 'hosting' + '.csv'         #Provedores de Cloud
input_shop = 'shop' + '.csv'               #Provedores de E-Commerce
input_server = 'Server.csv'                #Provedores de Server
input_payment = 'payment.csv'              #Provedores de Payments
input_shipping = 'shipping.csv'            #Provedores de Shipping

#----------------------------------------------------------------------------#
#--------------------CODIGOS DE TRATAMENTO DOS ARQUIVOS----------------------#
#----------------------------------------------------------------------------#

#ARQUIVO MX:
with open(input_file_path + input_mx, 'r') as _file:
    reader = csv.DictReader(_file)
    fieldnames = reader.fieldnames[1:]
    domains = []
    for row in reader:
        columns = []
        for column in fieldnames:
            if row[column] == 'x':
                columns.append(column)
        domains.append({'Dominio': row['Domain'], 'E-mail Provider': ','.join(columns)})
df_mx = pd.DataFrame(domains)

#ARQUIVO ANALYTICS:
with open(input_file_path + input_analytics, 'r') as _file:
    reader = csv.DictReader(_file)
    fieldnames = reader.fieldnames[1:]
    domains = []
    for row in reader:
        columns = []
        for column in fieldnames:
            if row[column] == 'x':
                columns.append(column)
        domains.append({'Dominio': row['Domain'], 'Analytics Providers': ','.join(columns)})
df_analytics = pd.DataFrame(domains)

#ARQUIVO CMS:
with open(input_file_path + input_cms, 'r') as _file:
    reader = csv.DictReader(_file)
    fieldnames = reader.fieldnames[1:]
    domains = []
    for row in reader:
        columns = []
        for column in fieldnames:
            if row[column] == 'x':
                columns.append(column)
        domains.append({'Dominio': row['Domain'], 'CMS Providers': ','.join(columns)})
df_cms = pd.DataFrame(domains)

#ARQUIVO FRAMEWORK:
with open(input_file_path + input_framework, 'r') as _file:
    reader = csv.DictReader(_file)
    fieldnames = reader.fieldnames[1:]
    domains = []
    for row in reader:
        columns = []
        for column in fieldnames:
            if row[column] == 'x':
                columns.append(column)
        domains.append({'Dominio': row['Domain'], 'Framework Providers': ','.join(columns)})
df_framework = pd.DataFrame(domains)

#ARQUIVO HOSTING:
with open(input_file_path + input_hosting, 'r') as _file:
    reader = csv.DictReader(_file)
    fieldnames = reader.fieldnames[1:]
    domains = []
    for row in reader:
        columns = []
        for column in fieldnames:
            if row[column] == 'x':
                columns.append(column)
        domains.append({'Dominio': row['Domain'], 'Hosting (Clouds) Providers': ','.join(columns)})
df_hosting = pd.DataFrame(domains)

#ARQUIVO SHOP:
with open(input_file_path + input_shop, 'r') as _file:
    reader = csv.DictReader(_file)
    fieldnames = reader.fieldnames[1:]
    domains = []
    for row in reader:
        columns = []
        for column in fieldnames:
            if row[column] == 'x':
                columns.append(column)
        domains.append({'Dominio': row['Domain'], 'E-Commerce Providers': ','.join(columns)})
df_shop = pd.DataFrame(domains)

#ARQUIVO Server:
with open(input_file_path + input_server, 'r') as _file:
    reader = csv.DictReader(_file)
    fieldnames = reader.fieldnames[1:]
    domains = []
    for row in reader:
        columns = []
        for column in fieldnames:
            if row[column] == 'x':
                columns.append(column)
        domains.append({'Dominio': row['Domain'], 'Server': ','.join(columns)})
df_server = pd.DataFrame(domains)



#ARQUIVO Payments:
with open(input_file_path + input_payment, 'r') as _file:
    reader = csv.DictReader(_file)
    fieldnames = reader.fieldnames[1:]
    domains = []
    for row in reader:
        columns = []
        for column in fieldnames:
            if row[column] == 'x':
                columns.append(column)
        domains.append({'Dominio': row['Domain'], 'Payment': ','.join(columns)})
df_payment = pd.DataFrame(domains)

#ARQUIVO Shipping:

#with open(input_file_path + input_shipping, 'r') as _file:
#    reader = csv.DictReader(_file)
#    fieldnames = reader.fieldnames[1:]
#    domains = []
#    for row in reader:
#        columns = []
#        for column in fieldnames:
#            if row[column] == 'x':
#                columns.append(column)
#        domains.append({'Dominio': row['Domain'], 'Shipping': ','.join(columns)})
#df_shipping = pd.DataFrame(domains)


#----------------------------------------------------------------------------#
#----------------------------------------------------------------------------#


#Cria um DataFrame final, onde unifica todos os dados de forma consolidada e reduz em uma Ãºnica planilha;
df_final = df_mx\
    .merge(df_analytics, how = 'outer')\
    .merge(df_cms, how = 'outer')\
    .merge(df_framework, how = 'outer')\
    .merge(df_hosting, how = 'outer')\
    .merge(df_shop, how = 'outer')\
    .merge(df_server, how = 'outer')\
    .merge(df_payment, how = 'outer')
    #.merge(df_shipping, how = 'outer')\
    #.merge(df_shipping, how = 'outer')


#Variaveis de OUTPUT;
#output_file_path = r'C:\Users\ian.rodrigues\OneDrive - IT Broker Servicos Operacionais LTDA (ISV)\Documentos\0.Demandas\4.MV\0.Enriquecimento\7.JOIN-5120-5123\8.B2W\input\\'
arquivo_output = 'output_tratamento_builtwith' + '.xlsx'

#Codigo para salvar;
df_final.to_excel(input_file_path + arquivo_output, index=False)

print('Lista exportada com sucesso! =)')
```
