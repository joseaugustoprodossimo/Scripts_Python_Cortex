```python
import pandas as pd
```

```python
caminho = 'C:\\Projetos\\Python\\DD Plus\\'
arquivo = '2 empresas'
df = pd.read_excel(caminho + arquivo + '.xlsx', converters={ 'Telefone 1': lambda x: str(x)})
```

```python
df['Capital Social até'] = df['Capital Social até'].str.replace(',', '')
df['Faturamento até'] = df['Faturamento até'].str.replace(',', '')
df['Quantidade de funcionários até'] = df['Quantidade de funcionários até'].str.replace(',', '')
```

```python
df['Capital Social até'] = pd.to_numeric(df['Capital Social até'])
df['Faturamento até'] = pd.to_numeric(df['Faturamento até'])
df['Quantidade de funcionários até'] = pd.to_numeric(df['Quantidade de funcionários até'])
```

```python
df['ESTADO'] = ''
df['SETOR'] = ''
df['TELEFONE ADMINISTRADOR'] = ''
df['KEYWORDS'] = ''
df['SEO DESCRIPTION'] = ''
df['COMPANY LINKEDIN URL'] = ''
df['FACEBOOK URL'] = ''
df['TWITTER URL'] = ''
df['TECHNOLOGIES'] = ''
df['TIPO LOGRADOURO'] = ''
```

```python
df['Telefone 1'] = df['Telefone 1'].str.replace('(', '').str.replace(')', '').str.replace('-', '').str.replace(' ', '')
df['TELEFONE'] = '+55' + df['Telefone 1']
```

```python
df['NOME DO ADMINISTRADOR'] = df['Nome do sócio 1']
```

```python
df['TIPO LOGRADOURO'] = df['Logradouro'].str.split(' ', expand=True).get(0)
df['Logradouro'] = df['Logradouro'].str.split(' ', expand=True).get(1)
```

```python
df.loc[df['UF'] == 'AC', 'ESTADO'] = 'ACRE'
df.loc[df['UF'] == 'AL', 'ESTADO'] = 'ALAGOAS'
df.loc[df['UF'] == 'AP', 'ESTADO'] = 'AMAPA'
df.loc[df['UF'] == 'AM', 'ESTADO'] = 'AMAZONAS'
df.loc[df['UF'] == 'BA', 'ESTADO'] = 'BAHIA'
df.loc[df['UF'] == 'CE', 'ESTADO'] = 'CEARA'
df.loc[df['UF'] == 'ES', 'ESTADO'] = 'ESPIRITO SANTO'
df.loc[df['UF'] == 'GO', 'ESTADO'] = 'GOIAS'
df.loc[df['UF'] == 'MA', 'ESTADO'] = 'MARANHAO'
df.loc[df['UF'] == 'MT', 'ESTADO'] = 'MATO GROSSO'
df.loc[df['UF'] == 'MS', 'ESTADO'] = 'MATO GROSSO DO SUL'
df.loc[df['UF'] == 'MG', 'ESTADO'] = 'MINAS GERAIS'
df.loc[df['UF'] == 'PA', 'ESTADO'] = 'PARA'
df.loc[df['UF'] == 'PB', 'ESTADO'] = 'PARAIBA'
df.loc[df['UF'] == 'PR', 'ESTADO'] = 'PARANA'
df.loc[df['UF'] == 'PE', 'ESTADO'] = 'PERNAMBUCO'
df.loc[df['UF'] == 'PI', 'ESTADO'] = 'PIAUI'
df.loc[df['UF'] == 'RJ', 'ESTADO'] = 'RIO DE JANEIRO'
df.loc[df['UF'] == 'RN', 'ESTADO'] = 'RIO GRANDE DO NORTE'
df.loc[df['UF'] == 'RS', 'ESTADO'] = 'RIO GRANDE DO SUL'
df.loc[df['UF'] == 'RO', 'ESTADO'] = 'RONDONIA'
df.loc[df['UF'] == 'RR', 'ESTADO'] = 'RORAIMA'
df.loc[df['UF'] == 'SC', 'ESTADO'] = 'SANTA CATARINA'
df.loc[df['UF'] == 'SP', 'ESTADO'] = 'SAO PAULO'
df.loc[df['UF'] == 'SE', 'ESTADO'] = 'SERGIPE'
df.loc[df['UF'] == 'TO', 'ESTADO'] = 'TOCANTINS'
df.loc[df['UF'] == 'DF', 'ESTADO'] = 'DISTRITO FEDERAL'
```

```python
df.drop(columns=['Capital Social'], inplace=True)
```

```python
df.rename(columns={'Código CNPJ': 'CNPJ'}, inplace= True)
df.rename(columns={'Razão social': 'RAZAO SOCIAL'}, inplace= True)
df.rename(columns={'Nome fantasia': 'NOME FANTASIA'}, inplace= True)
df.rename(columns={' Tipo situação': 'STATUS'}, inplace= True)
df.rename(columns={'Data da abertura': 'DATA ABERTURA'}, inplace= True)
df.rename(columns={'Logradouro': 'LOGRADOURO'}, inplace= True)
df.rename(columns={'Numero': 'NUMERO'}, inplace= True)
df.rename(columns={'Complemento': 'COMPLEMENTO'}, inplace= True)
df.rename(columns={'Cidade': 'MUNICIPIO'}, inplace= True)
df.rename(columns={'E-mail na Receita Federal': 'EMAIL'}, inplace= True)
df.rename(columns={'Código atividade econômica primária': 'CNAE PRINCIPAL CODIGO'}, inplace= True)
df.rename(columns={'Nome atividade econômica primária': 'CNAE PRINCIPAL DESCRICAO'}, inplace= True)
df.rename(columns={'Capital Social até': 'CAPITAL SOCIAL'}, inplace= True)
df.rename(columns={'Faturamento até': 'RECEITA ESTIMADA EMPRESA'}, inplace= True)
df.rename(columns={'Quantidade de funcionários até': 'NUMERO FUNCIONARIOS EMPRESA'}, inplace= True)
df.rename(columns={'Nome do sócio 1': 'SOCIO 1'}, inplace= True)
df.rename(columns={'Nome do sócio 2': 'SOCIO 2'}, inplace= True)
df.rename(columns={'Nome do sócio 3': 'SOCIO 3'}, inplace= True)
df.rename(columns={'Website1': 'WEBSITE'}, inplace= True)
```

```python
df = df.filter(items=[
    'CNPJ',
    'RAZAO SOCIAL',
    'NOME FANTASIA',
    'STATUS',
    'DATA ABERTURA',
    'TIPO LOGRADOURO',
    'LOGRADOURO',
    'NUMERO',
    'COMPLEMENTO',
    'BAIRRO',
    'CEP',
    'UF',
    'ESTADO',
    'MUNICIPIO',
    'TELEFONE',
    'EMAIL',
    'CNAE PRINCIPAL CODIGO',
    'CNAE PRINCIPAL DESCRICAO',
    'SETOR',
    'CAPITAL SOCIAL',
    'RECEITA ESTIMADA EMPRESA',
    'NUMERO FUNCIONARIOS EMPRESA',
    'SOCIO 1',
    'SOCIO 2',
    'SOCIO 3',
    'NOME DO ADMINISTRADOR',
    'TELEFONE ADMINISTRADOR',
    'WEBSITE',
    'KEYWORDS',
    'SEO DESCRIPTION',
    'COMPANY LINKEDIN URL',
    'FACEBOOK URL',
    'TWITTER URL',
    'TECHNOLOGIES'
])
```

```python
df.sort_values(by='RAZAO SOCIAL', axis=0, inplace=True)
```

```python
df.to_excel(caminho + 'Output ' + arquivo + '.xlsx', index= False)
```
