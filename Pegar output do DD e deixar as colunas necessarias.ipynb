```python
def colocar_ranges_padrao_DD(item):
  if item.between(0, 50000) or item.isnull():
    return "R$ 0 - R$ 50 K"
  elif item.between(50001, 250000):
    return "R$ 51 K - R$ 250 K"
  elif item.between(250001, 500000):
    return "R$ 251 K - R$ 500 K"
  elif item.between(500001, 2500000):
    return "R$ 501 K - R$ 2.5 M"
  elif item.between(2500001, 10000000):
    return "R$ 2.5 M - R$ 10 M"
  elif item.between(10000001, 50000000):
    return "R$ 10 M - R$ 50 M"
  elif item.between(50000001, 100000000):
    return "R$ 50 M - R$ 100 M"
  elif item.between(100000001, 250000000):
    return "R$ 100 M - R$ 250 MI"
  elif item.between(250000001, 500000000):
    return "R$ 250 M - R$ 500 MI"
  elif item.between(500000001, 1000000000):
    return "R$ 500 M - R$ 1 B"
  elif item > 1000000000:
    return "+ R$ 1 B"
  else:
      return item 
```
