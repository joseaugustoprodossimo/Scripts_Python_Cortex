```python
df = df.dropna(subset=["CNPJ"])
```

```python
df.loc[df['QUERY'].str.contains('/0001', regex=False), 'Tipo de unidade'] = 'Matriz'
df.loc[~df['QUERY'].str.contains('/0001', regex=False), 'Tipo de unidade'] = 'Filial'
```

```python
df.loc[df['CAPITAL SOCIAL'].between(0, 50000), 'FAIXA DE CAPITAL SOCIAL'] = "R$ 0 - R$ 50 K"
df.loc[df['CAPITAL SOCIAL'].between(50001, 250000), 'FAIXA DE CAPITAL SOCIAL'] = "R$ 51 K - R$ 250 K"
df.loc[df['CAPITAL SOCIAL'].between(250001, 500000), 'FAIXA DE CAPITAL SOCIAL'] = "R$ 251 K - R$ 500 K"
df.loc[df['CAPITAL SOCIAL'].between(500001, 2500000), 'FAIXA DE CAPITAL SOCIAL'] = "R$ 501 K - R$ 2.5 M"
df.loc[df['CAPITAL SOCIAL'].between(2500001, 10000000), 'FAIXA DE CAPITAL SOCIAL'] = "R$ 2.5 M - R$ 10 M"
df.loc[df['CAPITAL SOCIAL'].between(10000001, 50000000), 'FAIXA DE CAPITAL SOCIAL'] = "R$ 10 M - R$ 50 M"
df.loc[df['CAPITAL SOCIAL'].between(50000001, 100000000), 'FAIXA DE CAPITAL SOCIAL'] = "R$ 50 M - R$ 100 M"
df.loc[df['CAPITAL SOCIAL'].between(100000001, 250000000), 'FAIXA DE CAPITAL SOCIAL'] = "R$ 100 M - R$ 250 MI"
df.loc[df['CAPITAL SOCIAL'].between(250000001, 500000000), 'FAIXA DE CAPITAL SOCIAL'] = "R$ 250 M - R$ 500 MI"
df.loc[df['CAPITAL SOCIAL'].between(500000001, 1000000000), 'FAIXA DE CAPITAL SOCIAL'] = "R$ 500 M - R$ 1 B"
df.loc[df['CAPITAL SOCIAL'] > 1000000000, 'FAIXA DE CAPITAL SOCIAL'] = "+ R$ 1 B"
```

```python
df.loc[df['NUMERO FUNCIONARIOS EMPRESA'] == '0', 'NUMERO FUNCIONARIOS EMPRESA'] = '1 a 5'
```

```python
# CONCATENAR O DDD COM O TELEFONE
df['DDD'] = df['DDD'].str.replace('0', '')
df['Telefone 1'] = '+55' + df['DDD'] + df['TELEFONE']
```

```python
# SPLIT NOS SOCIOS
df['Nome do sócio 1'] = df['SOCIOS'].str.split(',', expand=True).get(1)
df['Nome do sócio 2'] = df['SOCIOS'].str.split(',', expand=True).get(2)
df['Nome do sócio 3'] = df['SOCIOS'].str.split(',', expand=True).get(3)
df['Nome do sócio 4'] = df['SOCIOS'].str.split(',', expand=True).get(4)
df['Nome do sócio 5'] = df['SOCIOS'].str.split(',', expand=True).get(5)
```

```python
df['WEBSITE'] = df['WEBSITE'].str.replace('http://', '').str.replace('https://', '').str.replace('www4.', '').str.replace('www2.', '').str.replace('www1.', '').str.replace('wwwa.', '').str.replace('www7.', '').str.replace('wwwt.', '').str.replace('wwws.', '').str.replace('www3.', '').str.replace('www5.', '')
df['WEBSITE'] = df['WEBSITE'].str.replace('www.', '')
df['WEBSITE'] = df['WEBSITE'].str.split('/', expand=True).get(0)
```

```python
df.rename(columns={'NOME FANTASIA': 'Nome fantasia',
                   'RAZAO SOCIAL': 'Razão social', 
                   'NUMERO': 'Numero', 
                   'COMPLEMENTO': 'Complemento', 
                   'BAIRRO': 'Bairro', 
                   'MUNICIPIO': 'Cidade', 
                   'DATA ABERTURA': 'Data da abertura', 
                   'STATUS': 'Tipo situação', 
                   'CNAE PRINCIPAL DESCRICAO': 'Nome atividade econômica primária', 
                   'CNAE PRINCIPAL CODIGO': 'Código atividade econômica primária', 
                   'EMAIL': 'E-mail na Receita Federal', 
                   'QUERY': 'Código CNPJ', 
                   'WEBSITE': 'Website1', 
                   'NUMERO FUNCIONARIOS EMPRESA': 'FAIXA DE FUNCIONÁRIOS', 
                   'RECEITA ESTIMADA EMPRESA': 'FAIXA DE FATURAMENTO'}, inplace= True)
```

```python
#Separa os sócios em DataFrames;
dfsocios1 = df.filter(items=['Código CNPJ','Razão social','Nome do sócio 1','Qualificação sócio 1','Tipo qualificação sócio 1'])
dfsocios2 = df.filter(items=['Código CNPJ','Razão social','Nome do sócio 2','Qualificação sócio 2','Tipo qualificação sócio 2'])
dfsocios3 = df.filter(items=['Código CNPJ','Razão social','Nome do sócio 3','Qualificação sócio 3','Tipo qualificação sócio 3'])
dfsocios4 = df.filter(items=['Código CNPJ','Razão social','Nome do sócio 4','Qualificação sócio 4','Tipo qualificação sócio 4'])
dfsocios5 = df.filter(items=['Código CNPJ','Razão social','Nome do sócio 5','Qualificação sócio 5','Tipo qualificação sócio 5'])

#Renomeia as colunas para padronizar e concatenar;
dfsocios1.rename(columns={'Código CNPJ':'CNPJ','Razão social':'RAZÃO SOCIAL','Nome do sócio 1':'NOME DO SÓCIO','Qualificação sócio 1':'QUALIFICAÇÃO SÓCIO', 'Tipo qualificação sócio 1':'TIPO QUALIFICAÇÃO SÓCIO'}, inplace=True)
dfsocios2.rename(columns={'Código CNPJ':'CNPJ','Razão social':'RAZÃO SOCIAL','Nome do sócio 2':'NOME DO SÓCIO','Qualificação sócio 2':'QUALIFICAÇÃO SÓCIO', 'Tipo qualificação sócio 2':'TIPO QUALIFICAÇÃO SÓCIO'}, inplace=True)
dfsocios3.rename(columns={'Código CNPJ':'CNPJ','Razão social':'RAZÃO SOCIAL','Nome do sócio 3':'NOME DO SÓCIO','Qualificação sócio 3':'QUALIFICAÇÃO SÓCIO', 'Tipo qualificação sócio 3':'TIPO QUALIFICAÇÃO SÓCIO'}, inplace=True)
dfsocios4.rename(columns={'Código CNPJ':'CNPJ','Razão social':'RAZÃO SOCIAL','Nome do sócio 4':'NOME DO SÓCIO','Qualificação sócio 4':'QUALIFICAÇÃO SÓCIO', 'Tipo qualificação sócio 4':'TIPO QUALIFICAÇÃO SÓCIO'}, inplace=True)
dfsocios5.rename(columns={'Código CNPJ':'CNPJ','Razão social':'RAZÃO SOCIAL','Nome do sócio 5':'NOME DO SÓCIO','Qualificação sócio 5':'QUALIFICAÇÃO SÓCIO', 'Tipo qualificação sócio 5':'TIPO QUALIFICAÇÃO SÓCIO'}, inplace=True)

#Concatena os dataframes num só;
df_socios = pd.concat([dfsocios1,dfsocios2,dfsocios3,dfsocios4,dfsocios5])

df_socios.dropna(subset=['NOME DO SÓCIO'], inplace=True)
```

```python
#Separa os sócios em DataFrames;
dfwebsite1 = df.filter(items=['Código CNPJ','Website1'])

#Renomeia as colunas para padronizar e concatenar;
dfwebsite1.rename(columns={'Código CNPJ':'CNPJ','Website1':'Website'}, inplace=True)

dfwebsite1.dropna(subset=['Website'], inplace=True)

Vazios = dfwebsite1[ dfwebsite1['Website'] == '' ].index

df.drop(Vazios , inplace=True)
```

```python
df = df.filter(items=['Código CNPJ',
                      'Razão social',
                      'Nome fantasia',
                      'Logradouro', 
                      'Numero',
                      'Complemento', 
                      'Bairro', 
                      'Cidade', 
                      'UF', 
                      'CEP', 
                      'Porte', 
                      'FAIXA DE CAPITAL SOCIAL', 
                      'FAIXA DE FATURAMENTO', 
                      'FAIXA DE FUNCIONÁRIOS', 
                      'Data da abertura',
                      'Tipo de unidade', 
                      'Tipo situação', 
                      'Nome atividade econômica primária', 
                      'E-mail na Receita Federal', 
                      'Telefone 1', 
                      'Telefone 2', 
                      'Telefone 3', 
                      'Telefone 4', 
                      'Telefone 5', 
                      'Website1', 
                      'Website2', 
                      'Website3', 
                      'Website4', 
                      'Website5'
])
```

```python
df['Nome fantasia'] = df['Nome fantasia'].str.title()
df['Cidade'] = df['Cidade'].str.upper()
```

```python
df.sort_values(by='Razão social', axis=0, inplace=True)
```

```python
#Cria um excel writer usando xlsxwriter como engine;
writer = pd.ExcelWriter(caminho + 'Output ' + arquivo + '.xlsx', engine='xlsxwriter')

#Escreve os dataframes em diferentes sheets dentro do arquivo final;
df.to_excel(writer, sheet_name='Dados Cadastrais - Cortex', index=False)
df_socios.to_excel(writer, sheet_name='Dados Sócios - Cortex', index=False)
dfwebsite1.to_excel(writer, sheet_name='Matriz CNPJ Website', index=False)

#Fecha o arquivo e salva no diretório;
writer.save()
```
